### <font color=darkcyan>Correlation beetween carbon/soil respiration/mineral in top soil vs bottom soil</font>

<strong>Soil copper and soil respiration</strong>

```python
sns.jointplot(x='soil_0_15_Cu', y='soil_15_30_Cu', data=df, kind='reg', stat_func=r2, height=8)
plt.subplots_adjust(top=1.2)
plt.xlabel('Copper in soil, 0-6" depths', fontsize=16)
plt.ylabel('Copper in soil, 6-12" depths', fontsize=16)
_ = plt.title('Copper in soil, 0 – 6” and 6 to 12” depths', fontsize=16, weight='bold')
```

There is a slight relationship between the copper content of the soil and the depth of the soil, the rate of Cu (copper) is nearly equivalent depending on the depth - The copper content of a soil does not depend on the geolocation of the soil but rather on the method and the agricultural activity: pesticides, composts, pig farming. So the copper accumulates in the soil and concentrates on the surface.

<strong>Soil Potassium and soil respiration</strong>

```python
sns.jointplot(x='soil_0_15_K', y='soil_15_30_K', data=df, stat_func=r2, kind='reg', height=8)
plt.subplots_adjust(top=1.2)
plt.xlabel('Potassium in soil, 0-6" depth', fontsize=16)
plt.ylabel('Potassium in soil, 6-12" depth', fontsize=16)
_ = plt.title('Potassium in soil, 0-6" and 6-12" depths', fontsize=16, weight='bold')
```

Potassium promotes plant development, as well as fruit and leaf/plant coloration, also making plants more resistant to disease. We can see that potassium is as much present on the soil surface as it is deep in the soil.

Based on the dataset, there is no obsvious relationship between the simple measurement of soil biological activity and any of the food quality measures. Graphs above show no obvious trends either by antioxidants, polyphenols, or mineral (as example, antioxidants by soil respiration).

## <font color=#F57D05>Carrot sample - Training models</font>

```python
# remove all missing values from the carrots dataframe
df_carrots_raw = pd.concat([df_carrotscan, df_target_carrots], axis=1 ,sort=False)
df_carrots_noNA = df_carrots_raw.dropna()
```

### <font color=#F57D05> Random Forest Classifier </font>

The random forest algorithm was tested with spectral data from the Bionutrient Meter (10 wavelengths from 365nm to 940nm).
<br>
We use Random Forest Classifier on purpose, not to predict the exact nutritional value of vegetables, but to determine whether a carrot (and spinach) is good or bad (in terms of nutrition). The goal is to determined whether a vegetable is good or bad compared to the median value of the nutrient, namely 32.73 for the antioxidant. Thus, a value below the threshold will be of class 0 (bad) and 1 for vegetables being above the threshold (good).

```python
# Minimally prepare dataset
# Independent variables
x = df_carrots_noNA[df_carrots_noNA.columns[:-3]].values
# variable to predict (antioxidant only because Random Forest Classifier can only take 1 output variable)
y = quantile(df_carrots_noNA[df_carrots_noNA.columns[-1:]].values)
```

```python
# We split the dataset into training set (to train the model) and testing set (to evaluate the model)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
print('Training Features Shape:', x_train.shape)
print('Training Labels Shape:', y_train.shape)
print('Testing Features Shape:', x_test.shape)
print('Testing Labels Shape:', y_test.shape)
```

```python
# Define the model
classifier = RandomForestClassifier(n_estimators=1000)
train_model(classifier, x_train, y_train)
```

```python
# predict whether the carrot is rich or poor in nutrient
# We get a score by looking where the model was right, how many times has the model correctly classified
score_classifier = scoreClassifier(classifier, x_test, y_test)
```

### <font color=#F57D05> Random Forest Regressor </font>
We use a regression model to try to predict the nutritional value of carrots in terms of polyphenol, protein and antioxidant based on spectral data.

```python
# Independent variables
x = df_carrots_noNA[df_carrots_noNA.columns[:-3]].values
# variables to predict (the 3 nutrients)
y = df_carrots_noNA[df_carrots_noNA.columns[-3:]].values
```

```python
# We split the dataset into training set and testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
print('Training Features Shape:', x_train.shape)
print('Training Labels Shape:', y_train.shape)
print('Testing Features Shape:', x_test.shape)
print('Testing Labels Shape:', y_test.shape)
```

```python
# Define the model
RF_regressor = RandomForestRegressor(n_estimators=1000, random_state=0)
# Train model with Random Forest Regressor
train_model(RF_regressor, x_train, y_train)
```

We compute the metrics to measure the performance of the model algorithm.
<br>
Then we display two plots : the 1st is to visualize the error made by the model on the 3 nutrients. The 2nd to see the distribution of the predicted of the predicted vales compared to the observed values.

```python
# evaluate model
# compute metrics
resume_metrics_rfr = compute_losses_reg(RF_regressor, x_test, y_test)

# ordinate axis labels
nutriments_units = ['Polyphenols (mg/100g)', 'Protein (mg/100g)', 'Antioxidant (µmol/100g)']
#plot mae and predicted variables
plot_mae_and_predict(model = RF_regressor,
                x = x_test,
                y = y_test,
                ylabel = nutriments_units)
```

We use a function that measures the performance of our model. We tried to determine how well our RandomForest Regressor model had predicted the nutritional values, by classifying the carrots as being low in nutrients or rich in nutrients (polyphenol, antioxidant, protein). For this, we established a threshold based on the 50th quantile of the observations values.

We can see the projection of the observation values and the predicted points for our three nutrients. The middle values form a block and more or less well predicted. Clouds of predicted values do not exactly match observed values, especially outliers of observed values which are more difficult to reach.<br>
Ideally, all our predicted points should be close to the observations in our three scatter plot. But here, we obtain  a low R2 for the RandomForest Regressor, which means that the model fit is weak.

### <font color=#F57D05> Linear Regression </font>
The linear regression algorithm was testedwith spectral data from the Bionutrient Meter (10 wavelengths from 365nm to 940nm). It is used to try to predict nutritional quality based on spectral data.

```python
# Define the model
lin_regression = LinearRegression()
train_model(lin_regression, x_train, y_train)
```

Here we compute the loss metrics of the linear regression algorithm on carrots testing set. So we obtain a dataframe including mae, mse, rmse, r2 and Score classifier metrics. We display the errors made by the model and the predicted values compared to the observed values

```python
# evaluate model
# compute metrics
resume_metrics_lin = compute_losses_reg(lin_regression, x_test, y_test)

# plot mae and predicted variables
plot_mae_and_predict(model = lin_regression,
                x = x_test,
                y = y_test,
                ylabel = nutriments_units)
```

### <font color=#F57D05>Comparison of the 3 learning algortihms on carrots dataset</font>

```python
# Report performances
df_resume_performance_carrot = pd.DataFrame.from_dict({
    'Linear Regression': resume_metrics_lin,
    'RandomForest Regressor': resume_metrics_rfr,
    'RandomForest Classifier': np.nan
    })
df_resume_performance_carrot['RandomForest Classifier'].loc['Score Classifier'] = score_classifier
df_resume_performance_carrot.fillna('', inplace=True)
df_resume_performance_carrot
```

we compare our 3 models to see if they can determine whether a carrot is rich in nutrients or not (polyphenols, protein, antioxidant). Our RandomForest Regressor describes 6% of the variations in the nutritional qualities of carrots in terms of polyphenols, antioxidants and proteins. 14% for Linear Regression. Which means that the predicted values does not match the observed values well (cf. plots above).

By comparing the results of the three models, we find that better performance are obtained using a RandomForest Classifier - to predict whether a carrot will be good or bad in nutritional value..<br>
We can see that our three learning algorithms (Randomforest Regressor,  Linear Regression and Random Forest Classifier) are not very efficient when we observe their coefficient of determination (R2). We could try different families of models and / or apply the regularization.<br>
Three types of models (random forest classifier, random forest regressor and linear regression) were used to correltate polyphenols, protein and antioxidants, with spectral data. We only used spectral data, that's may be why we couldn't avoid overfitting.

### <font color=green> Random Forest Regressor </font>

```python
x = df_spinach_noNA[df_spinach_noNA.columns[:-3]].values
y = df_spinach_noNA[df_spinach_noNA.columns[-3:]].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
print('Training Features Shape:', x_train.shape)
print('Training Labels Shape:', y_train.shape)
print('Testing Features Shape:', x_test.shape)
print('Testing Labels Shape:', y_test.shape)
```

```python
# Define the model
RF_regressor = RandomForestRegressor(n_estimators=1000, random_state=42)
train_model(RF_regressor, x_train, y_train)
```

```python
# evaluate model
# compute metrics and plot the errors made by the model and the predicted values compared to the observed values
resume_metrics_rfr = compute_losses_train_test_reg(model = RF_regressor,
                             x_train = x_train,
                             y_train = y_train,
                             x_test = x_test,
                             y_test = y_test)
resume_metrics_rfr.rename(columns={"Train": "RForest Regressor Train", "Test": "RForest Regressor Test"}, inplace=True)

#plot mae and predicted variables
plot_mae_and_predict(model = RF_regressor,
                x = x_test,
                y = y_test,
                ylabel = nutriments_units)
```

The nutritional values of sinpach are quite variable, therefore we see that the model adapts more or less well, in particular for polyphenol where the model predicts without necessarily matching with the values of observations. Same for the protein even if he has trouble reaching outliers. Even if the error is proportional to the range, there is a greater error in the antioxidant.

### <font color=green> Linear Regression </font>

```python
# Define the model
lin_regression = LinearRegression()
train_model(lin_regression, x_train, y_train)
```

```python
# evaluate model
# compute metrics and plot the errors made by the model and the predicted values compared to the observed values
resume_metrics_lin = compute_losses_train_test_reg(model = lin_regression,
                             x_train = x_train,
                             y_train = y_train,
                             x_test = x_test,
                             y_test = y_test)
resume_metrics_lin.rename(columns={"Train": "Linear Regression Train", "Test": "Linear Regression Test"}, inplace=True)

#plot mae and predicted variables
plot_mae_and_predict(model = lin_regression,
                x = x_test,
                y = y_test,
                ylabel = nutriments_units)
```

### <font color=green>Comparison of the 3 learning algorithms on spinach dataset</font>

```python
df_resume_performance_spinach = pd.concat([resume_metrics_lin, resume_metrics_rfr], axis=1)
df_resume_performance_spinach.insert(4, "Random Forest Classifier", [np.nan, np.nan, np.nan, np.nan, score_classifier])
df_resume_performance_spinach.fillna('', inplace=True)
df_resume_performance_spinach
```

Regarding training on the spinach dataset, we do not have enough data to allow the algorithm to understand the data and the model fails to generalize. This may explain the difference in results between the training set and the testing set for Linear Regression and Random Forest Regressor.

The models are learning well the form of the training set, however, it is not generalizing well, which explains (very) low and negative R2 on the testing set. This is an overfitting. This may be due to a very low number of data for training and testing phase.<br>For next time, to reduce overfitting it would be beneficial to try other approches :
- Add more data,
- Apply regularization,
- Features selection,
- PCA (reducing dimensionality of the data).
